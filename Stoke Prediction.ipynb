### Checking Continuous Variables for Outliers

```python
sns.boxenplot(data=df,x=df['age'])
plt.show()
```

```python
sns.boxenplot(data=df,x=df['avg_glucose_level'])
plt.show()
```

```python
sns.boxenplot(data=df,x=df['bmi'])
plt.show()
```

as we can see avg_glucose_level have some outliers and bmi have large number of outliers so we will eliminate them

#### Removing outliers in bmi

```python
Q1=df['bmi'].quantile(0.25)
```

```python
Q3=df['bmi'].quantile(0.75)
```

```python
IQR=Q3-Q1
```

```python
IQR
```

```python
LL=Q1-1.5*IQR
```

```python
UL=Q3+1.5*IQR
```

```python
LL
```

```python
UL
```

```python
df[df['bmi']<LL]
```

```python
df[df['bmi']>UL]
```

```python
df=df[(df['bmi']>LL) & (df['bmi']<UL)]
```

```python
df
```

#### Removing outliers for avg_glucose_level

```python
from scipy.stats import zscore
```

```python
df['zscore']=zscore(df['avg_glucose_level'])
```

```python
df
```

```python
df[df['zscore']<-3]
```

```python
df[df['zscore']>3]
```

```python
df=df[(df['zscore']>-3)&(df['zscore']<3)]
```

```python
df
```

```python
sns.boxenplot(data=df,x=df['bmi'])
plt.show()
```

```python
sns.boxenplot(data=df,x=df['avg_glucose_level'])
plt.show()
```

#### We reduced Outliers Sucessfully

## Dividing Training and Testing Dataset

```python
X=df.drop('stroke',axis=1).values
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(Y_test,Y_pred))
```

```python
print(classification_report(Y_test,Y_pred))
```

```python
LR.score(X_train,Y_train)
```
